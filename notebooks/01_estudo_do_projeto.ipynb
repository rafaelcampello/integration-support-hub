## O que é uma API?

API é uma interface para que sistemas conversem entre si. Em suporte a integrações, é comum analisar requisições, respostas, códigos HTTP, autenticação, logs e payloads.

## REST x SOAP

REST costuma usar HTTP e JSON de forma simples e flexível. SOAP é mais formal, baseado em XML, e aparece bastante em sistemas corporativos e legados.

Exemplo JSON REST:

```json
{
  "cliente_id": 123,
  "status": "ativo"
}
```

Exemplo XML SOAP simplificado:

```xml
<Envelope>
  <Body>
    <ConsultarCliente>
      <ClienteId>123</ClienteId>
    </ConsultarCliente>
  </Body>
</Envelope>
```

## Como funciona JWT

JWT é um token assinado enviado no cabeçalho `Authorization: Bearer <token>`. Ele permite que a API saiba quem está chamando sem pedir usuário e senha a cada requisição.

## Como o Swagger ajuda

O FastAPI gera documentação automática em `/docs`. Isso ajuda o time de suporte a descobrir endpoints, payloads esperados, respostas e códigos HTTP sem precisar ler todo o código.

## Estrutura do SQLite

- `users`: usuário padrão para autenticação.
- `integrations`: integrações REST/SOAP cadastradas.
- `integration_tests`: histórico de testes e falhas.
- `technical_logs`: eventos técnicos relevantes.

## Logs e troubleshooting

Logs ajudam a responder perguntas como: quem tentou autenticar, qual integração falhou, qual status foi retornado e se houve timeout. O projeto salva logs em arquivo e também registra eventos importantes no banco.

## Fluxo de teste de integração

1. A API busca a integração pelo ID.
2. Verifica se a integração está ativa.
3. Avalia se existe `failure_mode` para simular erro.
4. Executa o fluxo REST ou SOAP.
5. Salva resultado no histórico.
6. Registra evento técnico para análise posterior.

## Principais arquivos

- `app/main.py`: inicializa a aplicação.

- `app/models.py`: define tabelas.
- `app/services/integration_service.py`: concentra regras de teste.
